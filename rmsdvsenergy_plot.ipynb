```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
path = '/Users/da39/Desktop/VARANI_LAB/DA_cyrano_2019_FINAL_2/cyrano_simulations/'
file_name = 'RMSDVvsENERGY.txt'
```

```python
# listing all of the files in the directory
files = []
files = os.listdir(path)

# Getting rid of directories without RMSD files
files.remove('cyrano_default')
files.remove('figures')
```

```python

column_names = []
energy_rmsd_array = []

print (files)
for file in files:
    open_file = open(path + file + '/'+file_name).readlines()
    print ("working on file: %s" %(file))
    tmp_energy = []
    tmp_rmsd = []
    column_names.append("Energy")
    column_names.append("RMSD")
    line_counter = 0
    for line in open_file:           
            line_split = line.split(' ')
            line_split = [x for x in line_split if x != '']
            energy = float(line_split[1].rstrip('\n'))
            rmsd = float(line_split[0].rstrip('\n'))
            tmp_energy.append(energy)
            tmp_rmsd.append(rmsd)
            line_counter += 1;
    energy_rmsd_array.append(tmp_energy)
    energy_rmsd_array.append(tmp_rmsd)

        
```

```python
print (len(energy_rmsd_array))
```

```python
index_counter = 0;
for plots in range(0,len(energy_rmsd_array),2):
    print (plots) 
    plt.scatter(energy_rmsd_array[plots+1], energy_rmsd_array[plots])
    plt.title("Energy vs RMSD for " + str(files[index_counter]))
    plt.show()
    index_counter += 1;
```

```python

index_counter = 0;
for plots in range(0,len(energy_rmsd_array),2):
    print (plots) 
    fig = plt.figure(figsize=(20,10))
    
    #####################################
    #Plotting the ENERGY vs RMSD
    #####################################
    plt.subplot(2, 2, 2)
    plt.scatter(energy_rmsd_array[plots+1], energy_rmsd_array[plots], color='r', edgecolor='black', linewidth=1)
    plt.xlabel('RMSD', size='18')
    plt.ylabel('ENERGY', size='18')
    plt.title("Energy vs RMSD for " + str(files[index_counter]), size = '20')


    #####################################
    #Plotting the ENERGY 
    #####################################
    plt.subplot(2, 2, 1)
    plt.hist(energy_rmsd_array[plots], bins=20, color='r', edgecolor='black', linewidth=0.2, orientation='horizontal')
    #plt.hist(energy_new2*10, bins=20, color='r', edgecolor='black', linewidth=0.5)
    plt.xlabel('Counts', size='18')
    plt.ylabel('ENERGY', size='18')



    #####################################
    #Plotting the RMSD
    #####################################
    plt.subplot(2, 2, 4)
    plt.hist(energy_rmsd_array[plots+1], bins=30, color='r', edgecolor='black', linewidth=0.5)
    plt.xlabel('Counts', size='18')
    plt.ylabel('RMSD', size='18')
    plt.savefig('energy_rmsd_'+str(files[index_counter]) +'.png')
    plt.show()
    index_counter += 1;
    
    
```
